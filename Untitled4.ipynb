```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel('Mimi_andthegang_dataPreprocessing.xls')
df.head()
```

```python
plt.scatter(df[['Unnamed: 45']], df[['5. 5. Welchen monatlichen Beitrag sind Sie bereit zu zahlen, wenn die oben ausgewählten Eigenschaften in Ihrer Mitgliedschaft enthalten wären?']], color = "y", marker = "o", s = 40)
plt.xlabel('total numbers picked items') 
plt.ylabel('money spend for gym')
plt.title('Scatter Plot Title') 
plt.show()
```

```python
plt.scatter(df[['Unnamed: 46']], df[['5. 5. Welchen monatlichen Beitrag sind Sie bereit zu zahlen, wenn die oben ausgewählten Eigenschaften in Ihrer Mitgliedschaft enthalten wären?']], color = "y", marker = "o", s = 40)
plt.xlabel('total numbers picked items') 
plt.ylabel('money spend for gym')
plt.title('Scatter Plot Title') 
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

#define the classifier
classifier = LinearRegression()

#train the classifier
model = classifier.fit(df[['Unnamed: 45']], df['5. 5. Welchen monatlichen Beitrag sind Sie bereit zu zahlen, wenn die oben ausgewählten Eigenschaften in Ihrer Mitgliedschaft enthalten wären?'])

#use the trained classifier to make prediction
y_pred = classifier.predict(df[['Unnamed: 45']])
print(y_pred)

#print coefficient (a in y=ax+b) and intercept (the constant, b in y=ax+b)
print('Coefficients: \n', classifier.coef_)
print('Intercept: \n', classifier.intercept_)

#visualize original data points
plt.scatter(df[['Unnamed: 45']], df[['5. 5. Welchen monatlichen Beitrag sind Sie bereit zu zahlen, wenn die oben ausgewählten Eigenschaften in Ihrer Mitgliedschaft enthalten wären?']], color = "m", marker = "s", s = 30) 
#visualize regression function
plt.plot(df[['Unnamed: 45']], y_pred, color = "g") 
plt.xlabel('total numbers picked items') 
plt.ylabel('money spend for gym')
plt.title('Regression Function')
plt.show()
```
